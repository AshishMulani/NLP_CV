```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
plt.plot(np.arange(1, 40, 0.1), np.sin(np.arange(1, 40, 0.1)))
```

```python
image = cv2.imread('images/dog.jpg')
```

```python
rows, cols = image.shape[:2]
```

```python
rows, cols
```

```python
# create a grid

gridx, gridy = np.meshgrid(np.arange(cols), (np.arange(rows)))
```

```python
gridx, gridy
```

```python
# Define Mapping Rule

mapx = gridx
mapy = gridy + 25 * np.sin(gridx/20)
```

```python
mapx.dtype
```

```python
mapx = mapx.astype(np.float32)
mapy = mapy.astype(np.float32)
```

```python
mapped_img = cv2.remap(image, mapx, mapy, interpolation=cv2.INTER_LINEAR )
```

```python
mapped_img = cv2.cvtColor(mapped_img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(121); plt.title('Original Image')
plt.imshow(img);

plt.subplot(122); plt.title('Mapped Image')
plt.imshow(mapped_img);
```
