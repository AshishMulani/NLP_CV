```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('images/dog.jpg',0)
plt.imshow(img, cmap='gray')
plt.axis('off')
```

#### Binary Thresholding

```python
binary_img = img > 128
binary_img
```

```python
binary_img.dtype
```

```python
plt.imshow(binary_img, cmap = 'gray')
```

```python
img = cv2.imread('images/lenna.png',0)
binary_image = img > 128

plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(binary_image, cmap='gray')
```

```python
thresh, new_img = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)
thresh, new_img_inv = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY_INV)
```

```python
new_img.dtype
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,4,1);  plt.title('Original Image')
plt.imshow(img, cmap='gray')

plt.subplot(1,4,2); plt.title('Using Array Operations')
plt.imshow(binary_image, cmap='gray')

plt.subplot(1,4,3);  plt.title('Using Function')
plt.imshow(new_img, cmap='gray')

plt.subplot(1,4,4); plt.title('Inverse Binary')
plt.imshow(new_img_inv, cmap='gray')
```
