```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('images/lenna.png',0)
```

```python
help(cv2.Sobel)
```

```python
sobelx = cv2.Sobel(img , cv2.CV_64F , 1, 0 ,3)
sobely = cv2.Sobel(img , cv2.CV_64F , 0, 1 ,3)
```

```python
sobelx.dtype
```

```python
final_sobel = cv2.add(sobelx , sobely)
```

```python
thresh , binary_sobel = cv2.threshold(final_sobel , 60 ,255,cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1);plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(1,2,2);plt.title('Edges Image')
plt.imshow(binary_sobel, cmap='gray')
```

```python
cir_img = cv2.imread('images/circle.png',0)
```

```python
sobelx = cv2.Sobel(cir_img , cv2.CV_64F , 1, 0 ,3)
sobely = cv2.Sobel(cir_img , cv2.CV_64F , 0, 1 ,3)

thresh , sobelx_img = cv2.threshold(sobelx , 60 ,255, cv2.THRESH_BINARY)
thresh , sobely_img = cv2.threshold(sobely, 60 ,255, cv2.THRESH_BINARY)

final_sobel = cv2.add(sobelx , sobely)
thresh , binary_sobel = cv2.threshold(final_sobel , 60 ,255,cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,4,1);  plt.title('Original Image')
plt.imshow(cir_img, cmap='gray')

plt.subplot(1,4,2); plt.title('Sobel X')
plt.imshow(sobelx_img, cmap='gray')

plt.subplot(1,4,3);  plt.title('Sobel Y')
plt.imshow(sobely_img, cmap='gray')

plt.subplot(1,4,4); plt.title('Edges')
plt.imshow(binary_sobel, cmap='gray')
```

#### Canny Edge Detection

#### Used as pre-processing step in hand written digits and handwriting detection

```python
img = cv2.imread('images/lenna.png' , 0)
img_canny = cv2.Canny(img , 60 , 255)

plt.figure(figsize=(16,9))
plt.subplot(1,2,1);plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(1,2,2);plt.title('Edges Image')
plt.imshow(img_canny, cmap='gray')
```
