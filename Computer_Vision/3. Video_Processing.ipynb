##### Extract the frame / image at 42 second

```python
time = '00:00:42.00'
```

```python
hrs, mins, secs = time.split(':')
hrs, mins, secs
```

```python
fps = vid.get(cv2.CAP_PROP_FPS)
```

```python
tframe = int(hrs) * 3600 * fps + int(mins) * 60 * fps + float(secs) * fps 
```

```python
tframe
```

```python
vid.set(1, tframe)
```

```python
vid = cv2.VideoCapture(0)

# Get frame dimensions
frame_width = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 30.0, (frame_width, frame_height))

while True:
    ret, frame = vid.read()
    
    # Check if frame was read correctly
    if not ret or frame is None:
        print("Failed to grab frame. Exiting...")
        break
        
    cv2.imshow('My Video', frame)
    
    if cv2.waitKey(5) == ord('q'):
        break

vid.release()
out.release()
cv2.destroyAllWindows()
```
