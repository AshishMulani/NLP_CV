```python
img = cv2.imread('images/dog.jpg')
```

```python
kernel = np.zeros([3,3]) + 1/9
kernel
```

```python
blurred_img = cv2.filter2D(img, -1, kernel)
```

```python
cv2.imshow('Blurrred Image', blurred_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.zeros([9,9]) + 1/81
blurred_img = cv2.filter2D(img, -1, kernel)
img_new = np.hstack([img, blurred_img])

cv2.imshow('Blurrred Image', img_new)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Gaussian Kernel

```python
kernel_size =7
kernel = cv2.getGaussianKernel(kernel_size, 1.23)   #1.23 - Sigma(Standard Deviation value)
kernel
```

```python
gauss_img = cv2.filter2D(img, -1, kernel)
img_new = np.hstack([img, gauss_img])

cv2.imshow('Blurrred Image', img_new)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Median Blur

```python
img_gray = cv2.imread('images/dog.jpg', 0)
img_gray.shape
```

```python
img_med = cv2.medianBlur(img_gray, 7)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.imshow(img_gray, cmap = 'gray')
plt.subplot(122)
plt.imshow(img_med, cmap = 'gray')
```

```python
img_gray[100:200, 150:200] = cv2.medianBlur(img_gray[100:200, 150:200], 9)

plt.figure(figsize=(10,6))
plt.imshow(img_gray, cmap = 'gray')
```
