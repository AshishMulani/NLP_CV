```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/dog.jpg', 0)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(img, cmap='Greys')
```

```python
orig_map = plt.colormaps['Greys'].reversed()

# Plotting the comparison
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')      # Standard grayscale
plt.subplot(1, 2, 2)
plt.imshow(img, cmap=orig_map)    # Reversed grayscale (negative image)
```

```python
plt.figure(figsize=(16, 16))
plt.subplot(441)
plt.imshow(img, cmap=plt.colormaps['Greys'].reversed()); plt.title('Greys')
plt.subplot(442)
plt.imshow(img, cmap=plt.colormaps['Blues'].reversed()); plt.title('Blues')
plt.subplot(443)
plt.imshow(img, cmap=plt.colormaps['Reds'].reversed()); plt.title('Reds')
plt.subplot(444)
plt.imshow(img, cmap=plt.colormaps['Greens'].reversed()); plt.title('Greens')
plt.figure(figsize=(16, 16))
plt.subplot(445)
plt.imshow(img, cmap=plt.colormaps['Purples'].reversed()); plt.title('Purples')
plt.subplot(446)
plt.imshow(img, cmap=plt.colormaps['Oranges'].reversed()); plt.title('Oranges')
plt.subplot(447)
plt.imshow(img, cmap=plt.colormaps['PuRd'].reversed()); plt.title('PuRd')
plt.subplot(448)
plt.imshow(img, cmap=plt.colormaps['BuPu'].reversed()); plt.title('BuPu')
```

#### Diverging ColorMaps

```python
plt.figure(figsize=(16,16))
plt.subplot(441)
plt.imshow(img, cmap='PiYG'); plt.title('PiYG')
plt.subplot(442)
plt.imshow(img, cmap=plt.colormaps['PRGn'].reversed()); plt.title('PRGn')
plt.subplot(443)
plt.imshow(img, cmap=plt.colormaps['BrBG'].reversed()); plt.title('BrBG')
plt.subplot(444)
plt.imshow(img, cmap=plt.colormaps['PuOr'].reversed()); plt.title('PuOr')
plt.subplot(445)
plt.imshow(img, cmap=plt.colormaps['RdGy'].reversed()); plt.title('RdGy')
plt.subplot(446)
plt.imshow(img, cmap=plt.colormaps['coolwarm'].reversed()); plt.title('coolwarm')
plt.subplot(447)
plt.imshow(img, cmap=plt.colormaps['Spectral'].reversed()); plt.title('Spectral')
plt.subplot(448)
plt.imshow(img, cmap=plt.colormaps['bwr'].reversed()); plt.title('bwr')
```

#### Cyclic ColorMaps

```python
plt.figure(figsize=(16, 16))
plt.subplot(441)
plt.imshow(img, cmap='twilight'); plt.title('twilight')
plt.subplot(442)
plt.imshow(img, cmap='twilight_shifted'); plt.title('twilight_shifted')
plt.subplot(443)
plt.imshow(img, cmap='hsv'); plt.title('hsv')
plt.subplot(444)
plt.imshow(img, cmap='viridis'); plt.title('viridis')
```
