```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image = plt.imread('images/cat_damaged.png')
plt.imshow(image)
```

```python
image.shape
```

```python
height, width, _ = image.shape
```

```python
height, width
```

```python
mask = np.zeros([height, width, 3], dtype=np.uint8)
plt.imshow(mask, cmap='gray')
```

```python
for i in range(height):
    for j in range(width):
        if image[i,j].sum() == 0:
            mask[i,j] = [255,255,255]
```

```python
plt.imshow(mask, cmap='gray')
```

```python
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.subplot(122)
plt.imshow(mask, cmap='gray')
```

```python
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
```

```python
## Inpaint

image = cv2.imread('images/cat_damaged.png')
dst = cv2.inpaint(image, mask, 3, cv2.INPAINT_NS)
```

```python
plt.subplot(121); plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB));

plt.subplot(122); plt.title('Reconstructed Image')
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB));
```
