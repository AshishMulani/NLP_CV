```python
lines = [
    'It was a nice rainy day.',
    'The things were so beautiful on his point.',
    'When your focus is clear, you won.',
    'Many may happy returns of the day!',
    'One book can change your life.'
]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(lines)
```

```python
tokenizer.word_docs
```

```python
tokenizer.word_counts
```

```python
tokenizer.word_index
```

```python
tokenizer.index_word
```

```python
tokenizer.index_docs
```

```python
mat = tokenizer.texts_to_matrix(lines)
mat
```

```python
mat.shape
```

```python
tokenizer.word_index    
```

```python
words_in_first_row = [
    tokenizer.index_word[i] for i, value in enumerate(mat[0])
    if value == 1.0 and i != 0
]
```

```python
words_in_first_row
```

```python
words_in_second_row = [
    tokenizer.index_word[i] for i, value in enumerate(mat[1])
    if value == 1.0 and i != 0
]
words_in_second_row
```

```python
# 1. Invert the word index (index -> word)
index_to_word = {index: word for word, index in tokenizer.word_index.items()}

# 2. Get the first row from the matrix
first_sentence_row = mat[0]

# 3. Extract words where the matrix has a 1
first_sentence_words = [
    index_to_word[i] 
    for i, value in enumerate(first_sentence_row) 
    if value == 1 and i in index_to_word
]

print(first_sentence_words)
```

```python
seq = tokenizer.texts_to_sequences(lines)
seq
```

```python
padded = pad_sequences(seq, maxlen=10,padding='post')
padded
```

```python
padded = pad_sequences(seq, maxlen=10,padding='pre')
padded
```
