```python

alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

char_to_num = {}

for i in range(len(alphabets)):
    letter = alphabets[i]
    char_to_num[letter] = i

print(char_to_num)
```

```python
int_to_char = dict(enumerate(alphabets))
int_to_char
```

```python
char_to_int = dict((v, i) for i, v in enumerate(alphabets))
char_to_int
```

#### prepare the data

```python
# Assuming 'alphabets' is already defined: alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
seq_length = 1
x = []
y = []

# Populate x and y
for i in range(len(alphabets) - seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i + 1]
    
    print(seq_in, '-->', seq_out)
    
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
```

```python
x
```

```python
y
```

```python
import numpy as np
```

```python
x =np.reshape(x, (25,1,1))
x
```

```python
x_scaled = x/25
x_scaled
```

```python
from keras.utils import to_categorical
```

```python
y_new = to_categorical(y)
y_new
```

```python
y_new.shape
```

#### Build the model

```python
from keras.layers import SimpleRNN, Input, Dense
from keras.models import Sequential
```

```python
model = Sequential()

model.add(Input((1,1)))
model.add(Dense(32, activation = 'relu'))
model.add(SimpleRNN(32))
model.add(Dense(26, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer ='adam', metrics=['accuracy'])
```

```python
model.fit(x_scaled, y_new, batch_size=1, epochs=1000)
```

```python
model.evaluate(x_scaled, y_new, batch_size=1)
```

```python
# Assuming 'alphabets' is already defined: alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
seq_length = 3
x = []
y = []

# Populate x and y
for i in range(len(alphabets) - seq_length):
    seq_in = alphabets[i:i+seq_length]
    seq_out = alphabets[i + seq_length]
    
    print(seq_in, '-->', seq_out)
    
    x.append([char_to_int[char] for char in seq_in])
    y += [char_to_int[seq_out]]
```

```python
x_new = np.reshape(x, (23, seq_length, 1))
```

```python
y_new = to_categorical(y)
```

```python
x_new.shape, y_new.shape
```

```python
x_new = x_new / 25
```

```python
model = Sequential()

model.add(Input((3,1)))
model.add(Dense(32, activation = 'relu'))
model.add(SimpleRNN(32))
model.add(Dense(26, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer ='adam', metrics=['accuracy'])
```

```python
model.fit(x_new, y_new, batch_size=1, epochs=1000)
```

```python
model.evaluate(x_new, y_new, batch_size=1)
```
