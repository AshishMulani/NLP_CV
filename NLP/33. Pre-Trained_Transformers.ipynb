```python
from transformers import pipeline
```

#### Text Classification

```python
classifier = pipeline('text-classification')
```

```python
import pandas as pd
```

```python
text = 'It was possible for him. So he did it.'
```

```python
result = classifier(text)
```

```python
df = pd.DataFrame(result)
```
