```python
import pyiwn
```

```python
iwn = pyiwn.IndoWordNet(lang=pyiwn.Language.MARATHI)
```

```python
word = iwn.synsets('आंबा')
word
```

```python
word = iwn.synsets('धरती')
word
```

```python
# Explore the specific synsets

word[0].synset_id()
```

```python
word[0].pos()
```

```python
word[0].gloss() # meaning
```

```python
word[0].examples() # example in sentence
```

```python
word[0].lemma_names() # synonyms
```

## Hyponyms

```python
hyponyms = iwn.synset_relation(word[0], pyiwn.SynsetRelations.HYPONYMY)
```

```python
for h in hyponyms :
    print(h.lemma_names()[0])
```

```python
word = iwn.synsets('आंबा')
hyponyms = iwn.synset_relation(word[0], pyiwn.SynsetRelations.HYPONYMY)
for h in hyponyms :
    print(h.lemma_names()[0])
```
