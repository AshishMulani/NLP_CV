```python
nlp = spacy.load('en_core_web_sm')
nlp
```

```python
sent = nlp(''' Mark Zukerburg will meet Aditya Joshi on Monday 25 November 2015, 10 am for $3 Trillion deal at Pune for second time. ''')
sent
```

```python
sent.ents
```

```python
for ent in sent.ents:
    print(ent.text,'---->', ent.label_)
```
