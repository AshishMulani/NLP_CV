## Finding Top 10 Words Occured Frequently in This Article

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
```

```python
from nltk import FreqDist
freq = FreqDist(tokens)
freq.plot(10)
```

```python
from nltk.corpus import stopwords
```

```python
tokens1 = [token.lower() for token in tokens if token.isalpha() and token.lower() not in stopwords.words('english')]
freq = FreqDist(tokens1)
freq.plot(20)
```

```python
freq.most_common(20)
```

### Applying Stemming

```python
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer
ls = LancasterStemmer()
```

```python
tokens1 = [ls.stem (token.lower()) for token in tokens if token.isalpha() and token.lower() not in stopwords.words('english')]
freq = FreqDist(tokens1)
freq.plot(20)
```

```python
freq.most_common(20)
```
