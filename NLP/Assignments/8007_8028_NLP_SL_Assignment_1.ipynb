```python
import re
import spacy
import time

# 1. Messy text
text = """
John visited New York on 12/05/2024.
Meeting held in London on 2025-01-15.
Random text ### Pune office shifted on 18-08-2023.
Next event at Mumbai on 03/11/2026.
"""

# 2. Rule-based extraction (Regex)
start = time.time()

date_pattern = r'\b(?:\d{2}[/-]\d{2}[/-]\d{4}|\d{4}-\d{2}-\d{2})\b'
dates_regex = re.findall(date_pattern, text)

location_pattern = r'\b(?:New York|London|Pune|Mumbai)\b'
locations_regex = re.findall(location_pattern, text)

regex_time = time.time() - start

# 3. spaCy NER
nlp = spacy.load("en_core_web_sm")

start = time.time()

doc = nlp(text)

dates_spacy = [ent.text for ent in doc.ents if ent.label_ == "DATE"]
locations_spacy = [ent.text for ent in doc.ents if ent.label_ in ["GPE", "LOC"]]

spacy_time = time.time() - start

# 4. Results
print("=== REGEX ===")
print("Dates:", dates_regex)
print("Locations:", locations_regex)
print("Time:", round(regex_time, 6), "seconds")

print("\n=== SPACY NER ===")
print("Dates:", dates_spacy)
print("Locations:", locations_spacy)
print("Time:", round(spacy_time, 6), "seconds")

# 5. Comparison
print("\n=== COMPARISON ===")
print("Regex: Faster, but only finds patterns explicitly defined.")
print("spaCy: Slower, but can detect entities without predefined rules.")

# 6. Conclusion
print("\nConclusion: Use Regex for fixed formats and spaCy NER for flexible real-world text.")
```

###### • Word Sense Disambiguation in Practice: 1. Create a simple script using the Lesk algorithm from the NLTK library to disambiguate polysemous words, such as "bank" or "bat," across different contextual sentences. Analyze the output to identify specific edge cases where the algorithm fails to capture the correct semantic meaning.

```python
# 3.Word Sense Disambiguation in Practice:
import nltk
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

sentences = [
    ("I deposited money in the bank yesterday.", "bank"),
    ("The fisherman sat on the bank of the river.", "bank"),
    ("The bat flew out of the cave at night.", "bat"),
    ("He hit the ball with a bat.", "bat"),
    ("The bank approved my loan near the river.", "bank"),  # edge case
    ("The bat was kept beside the wicket as a bat flew overhead.", "bat")  # edge case
]

print("=== LESK WORD SENSE DISAMBIGUATION ===\n")

for sentence, target_word in sentences:
    tokens = word_tokenize(sentence)
    sense = lesk(tokens, target_word)

    print("Sentence:", sentence)
    print("Target Word:", target_word)

    if sense:
        print("Predicted Sense:", sense.name())
        print("Definition:", sense.definition())
    else:
        print("No sense found")

    print("-" * 60)

print("\n=== ANALYSIS ===")
print("1. Lesk usually works when strong context words are present (money, river, cave, ball).")
print("2. It compares dictionary definitions with surrounding words and selects the best overlap.")
print("3. Edge Case: 'The bank approved my loan near the river.'")
print("   - Contains clues for both financial bank and river bank.")
print("   - Lesk may choose the wrong sense because both meanings appear in the context.")
print("4. Edge Case: 'The bat was kept beside the wicket as a bat flew overhead.'")
print("   - Two different meanings of 'bat' occur in the same sentence.")
print("   - Lesk predicts only one sense and cannot distinguish both occurrences separately.")
print("5. Lesk relies on dictionary word overlap, so it often struggles with short, ambiguous, or complex sentences.")
```

###### • Building Domain-Specific Ontologies: 1. Utilize a no-code or low-code graphical ontology editor, such as WebProtégé, to design a small knowledge graph for a specific domain like an educational institution. Write a basic script to parse university website text and map the extracted entities to the relationships defined in your custom ontology.

```python
# 4.Building Domain-Specific Ontologies:

import spacy

nlp = spacy.load("en_core_web_sm")

text = """
SPPU University has a Computer Science Department.
Professor Tushar Sir teaches Natural Language Processing.
Professor Sunita Sharma teaches Machine Learning.
Professor Amit Joshi teaches Advanced Web Technologies.
Students are enrolled in Natural Language Processing and Machine Learning and Advanced Web Technologies.
"""

doc = nlp(text)

university = "SPPU University"
department = "Computer Science Department"

triples = []

triples.append((university, "hasDepartment", department))

# Extract professors
professors = []

for ent in doc.ents:
    if ent.label_ == "PERSON":
        professors.append(ent.text)

# Map teaching relationships
sentences = list(doc.sents)

for sent in sentences:
    s = sent.text

    if "teaches" in s:
        teacher = None

        for prof in professors:
            if prof in s:
                teacher = prof

        course = s.split("teaches")[-1].strip().replace(".", "")

        if teacher:
            triples.append((teacher, "teaches", course))
            triples.append((department, "offersCourse", course))

# Student enrollment
for sent in sentences:
    s = sent.text

    if "enrolled in" in s:
        courses = s.split("enrolled in")[-1]
        courses = courses.replace(".", "")
        courses = [c.strip() for c in courses.split(" and ")]

        for course in courses:
            triples.append(("Student", "enrolledIn", course))

# Output knowledge graph triples
print("=== KNOWLEDGE GRAPH TRIPLES ===\n")

for subj, rel, obj in triples:
    print(f"({subj}) --{rel}--> ({obj})")

print("\n=== CONCLUSION ===")
print("Entities extracted from website text were mapped to ontology-defined relationships.")
print("The resulting triples form a simple knowledge graph that can be extended in WebProtégé.")
```

###### • Traditional Machine Learning versus Zero-Shot Classification: 1. Train a basic Naive Bayes or Support Vector Machine classifier for text categorization using a standard movie review dataset. Process the exact same dataset using a zero-shot classification pipeline from Hugging Face, comparing the traditional model's training time against the transformer's out-of-the-box performance and interpretability.

```python
# 5.Traditional Machine Learning versus Zero-Shot Classification:
import nltk
import random
import time

from nltk.corpus import movie_reviews
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy

from transformers import pipeline

nltk.download('movie_reviews')

documents = [
    (list(movie_reviews.words(fileid)), category)
    for category in movie_reviews.categories()
    for fileid in movie_reviews.fileids(category)
]

random.shuffle(documents)

# Feature extraction
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())
word_features = list(all_words.keys())[:2000]

def document_features(document):
    words = set(document)
    return {word: (word in words) for word in word_features}

featuresets = [(document_features(d), c) for (d, c) in documents]

split = int(len(featuresets) * 0.8)
train_set = featuresets[:split]
test_set = featuresets[split:]

# Train Naive Bayes
start = time.time()

classifier = NaiveBayesClassifier.train(train_set)

nb_train_time = time.time() - start
nb_accuracy = accuracy(classifier, test_set)

print("=== NAIVE BAYES ===")
print("Training Time:", round(nb_train_time, 2), "seconds")
print("Accuracy:", round(nb_accuracy * 100, 2), "%")

print("\nMost Informative Features:")
classifier.show_most_informative_features(5)

#  Zero-Shot Classification
classifier_zs = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)

candidate_labels = ["positive", "negative"]

test_reviews = []
true_labels = []

for i in range(20):  # small sample to keep runtime reasonable
    words, label = documents[split + i]
    text = " ".join(words[:300])  # truncate long reviews
    test_reviews.append(text)
    true_labels.append(label)

start = time.time()

correct = 0

for review, true_label in zip(test_reviews, true_labels):
    result = classifier_zs(review, candidate_labels)

    predicted = result["labels"][0]

    if predicted == true_label:
        correct += 1

zs_time = time.time() - start
zs_accuracy = correct / len(test_reviews)

print("\n=== ZERO-SHOT TRANSFORMER ===")
print("Inference Time:", round(zs_time, 2), "seconds")
print("Accuracy:", round(zs_accuracy * 100, 2), "%")

# 4. Comparison
print("\n=== COMPARISON ===")
print("Naive Bayes:")
print("- Requires training.")
print("- Very fast training and prediction.")
print("- Easy to interpret via informative features.")

print("\nZero-Shot Transformer:")
print("- No task-specific training required.")
print("- Much slower inference.")
print("- Better semantic understanding.")
print("- Harder to interpret internally.")

print("\nConclusion:")
print("Naive Bayes is lightweight, fast, and interpretable. Zero-shot transformers require no training and generalize better, but are computationally expensive and less interpretable.")
```

###### • Investigating Bias in Word Embeddings: 1. Download pre-trained GloVe or Word2Vec vectors and write a Python script utilizing cosine similarity to solve word analogies. Actively search for and document historical, cultural, or gender biases present in the nearest- neighbor results, reflecting on how training data influences semantic representations.

```python
# 6.Investigating Bias in Word Embeddings:
import numpy as np

embeddings = {
    "man":       np.array([-1.0,  0.0,  0.0]),
    "king":      np.array([-0.9,  0.9,  0.0]),
    "woman":     np.array([ 1.0,  0.0,  0.0]),
    "queen":     np.array([ 0.9,  0.9,  0.0]),
    
    "paris":     np.array([ 0.0,  0.5,  1.0]),
    "france":    np.array([ 0.0,  0.0,  1.2]),
    "italy":     np.array([ 0.0,  0.0,  1.5]),
    "rome":      np.array([ 0.0,  0.5,  1.3]),
    
    "boy":       np.array([-1.0, -0.5,  0.0]),
    "girl":      np.array([ 1.0, -0.5,  0.0]),
    "brother":   np.array([-1.0, -0.2,  0.0]),
    "sister":    np.array([ 1.0, -0.2,  0.0]),
    
    "doctor":    np.array([-0.4,  0.8,  0.1]),
    "programmer":np.array([-0.6,  0.7,  0.1]),
    "nurse":     np.array([ 0.5,  0.6,  0.1]),
    "homemaker": np.array([ 0.7, -0.1,  0.1]),
    
    "physician": np.array([-0.35, 0.8,  0.1]),
    "male":      np.array([-0.95, 0.0,  0.0]),
    "coder":     np.array([-0.55, 0.7,  0.1]),
    "female":    np.array([ 0.95, 0.0,  0.0]),
    "registered":np.array([ 0.45, 0.6,  0.1]),
    "housewife": np.array([ 0.75,-0.1,  0.1]),
}

def cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

def most_similar_mock(positive, negative=[], exclude=[], topn=1):
    target_vector = np.zeros(3)
    for p in positive: target_vector += embeddings[p]
    for n in negative: target_vector -= embeddings[n]
    
    scores = []
    for word, vec in embeddings.items():
        if word in positive or word in negative or word in exclude:
            continue
        scores.append((word, cosine_similarity(target_vector, vec)))
        
    scores.sort(key=lambda x: x[1], reverse=True)
    return scores[:topn]

# 2. Word Analogies (man : king :: woman : ?)
print("=== WORD ANALOGIES ===")
analogies = [
    ("man", "king", "woman"),
    ("paris", "france", "italy"),
    ("boy", "girl", "brother")
]

for a, b, c in analogies:
    result = most_similar_mock(positive=[c, b], negative=[a], topn=1)[0]
    print(f"{a} : {b} :: {c} : {result[0]} (Similarity: {result[1]:.3f})")

# 3. Bias Exploration
print("\n=== BIAS EXPLORATION ===")
bias_words = ["doctor", "nurse", "programmer", "homemaker"]

for word in bias_words:
    print(f"\nNearest neighbors of '{word}':")
    neighbors = most_similar_mock(positive=[word], exclude=[word], topn=2)
    for w, score in neighbors:
        print(f"  {w:12s} {score:.3f}")
```

###### • Visualizing Transformer Self-Attention: 1. Access a free online visualization tool for transformer architectures, such as BertViz. Input a complex sentence containing ambiguous pronouns and trace the self-attention weights across different layers to observe exactly which surrounding words the model focuses on to resolve the syntactic ambiguity.

```python
# 7.Visualizing Transformer Self-Attention:
from transformers import AutoTokenizer, AutoModel
from bertviz import head_view

# Ambiguous sentence
sentence = "John told Michael that he would receive the award because he had worked very hard."

# Load model
model_name = "bert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name, output_attentions=True)

# Tokenize
inputs = tokenizer(sentence, return_tensors="pt")

# Get attention weights
outputs = model(**inputs)

attention = outputs.attentions
tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])

# Launch BertViz visualization
head_view(attention, tokens)
```

###### • Rapid Low-Code Model Deployment: 1. Construct a lightweight web application for text summarization or sentiment analysis by wrapping a Hugging Face pipeline in a Gradio or Streamlit interface. This exercise demonstrates how to quickly make complex NLP models accessible and interactive for non-technical end users without writing extensive backend code.

```python
# 8.Rapid Low-Code Model Deployment:
import gradio as gr
from transformers import pipeline

# 1. Load Hugging Face model
classifier = pipeline("sentiment-analysis")

# 2. Prediction function
def analyze_sentiment(text):
    result = classifier(text)[0]
    return f"Sentiment: {result['label']}\nConfidence: {result['score']:.4f}"

# 3. Gradio Interface
app = gr.Interface(
    fn=analyze_sentiment,
    inputs=gr.Textbox(lines=5, placeholder="Enter text here..."),
    outputs="text",
    title="Sentiment Analysis App",
    description="Analyze sentiment using a Hugging Face Transformer model."
)

# 4. Launch
app.launch()
```

###### • Acoustic Challenges in Speech Processing: 1. Record multiple short audio clips of the same sentence, varying the background noise levels and speaking styles. Process these recordings through an open-source speech-to-text system like OpenAI's Whisper, calculating the Word Error Rate for each transcription to understand the practical limitations of acoustic models in real-world environments.

```python
import whisper
from jiwer import wer
import os

# 1. Reference sentence
reference = "Natural language processing is transforming how humans interact with computers"

# 2. Audio files (Update these paths to match your actual local Windows folders)
# Use 'r' before the string to handle backslashes correctly on Windows
audio_files = [
    r"C:/Users/PGCP-AI/audio-samples/ComfyUI_00001_.mp3",
    r"C:/Users/PGCP-AI/audio-samples/ComfyUI_00002_.mp3",
    r"C:/Users/PGCP-AI/audio-samples/ComfyUI_00003_.mp3",
]

# 3. Load Whisper model
model = whisper.load_model("base")

print("=== WHISPER SPEECH RECOGNITION ===\n")

for audio in audio_files:
    # Check if the file exists before attempting to process
    if not os.path.exists(audio):
        print(f"Error: File not found at {audio}")
        continue
        
    print(f"Processing: {audio}")
    result = model.transcribe(audio)
    transcription = result["text"].strip()

    error_rate = wer(reference.lower(), transcription.lower())

    print("Transcription:", transcription)
    print("WER:", round(error_rate, 3))
    print("-" * 60)

print("\n=== ANALYSIS ===")
print("1. Clean audio generally produces the lowest WER.")
print("2. Increasing background noise usually increases transcription errors.")
print("3. Fast speech may cause word omissions or substitutions.")
print("4. Slow speech is often recognized more accurately.")
print("5. Whisper is robust but performance degrades in challenging acoustic environments.")
```
