```python
import nltk
from nltk import word_tokenize, pos_tag
from nltk.stem import WordNetLemmatizer
from bs4 import BeautifulSoup
import requests
from collections import Counter
```

```python
# Open the wikipedia page https://en.wikipedia.org/wiki/Saturn

url = 'https://en.wikipedia.org/wiki/Saturn'

headers = {
    'User-Agent': 'Custom-Script/1.9 (tushar@gmail.com)',
    'Accept-Language': 'en-US,en;q=0.5'
}
```

```python
response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
data = soup.get_text()
print(data)
```

```python
# 1. Scrap the web and find top 10 nouns from the page

clean = [word for word in word_tokenize(data) if word.isalpha() or word.isdigit()]
tags = nltk.pos_tag(clean)
```

```python
nouns = [word for word, tag in tags if tag.startswith('NN')]
```

```python
noun_counts = Counter(nouns)
for word, count in noun_counts.most_common(10):
    print(word, count)
```

```python
# 2. Print all the unique verbs in their root form.

lem = WordNetLemmatizer()

for word, tag in tags:
    if tag.startswith('V'):
        root_word = lem.lemmatize(word.lower(), pos='v')
        print(f"{word} --> {root_word}")
```

```python
# 3. Extract all the <DT><JJ><NN> phrases from the article.

pairs =[]
for i in range(len(tags)-2):
    if (tags[i][1].startswith('DT') and 
        tags[i+1][1].startswith('JJ') and 
        tags[i+2][1].startswith('NN')):
        
        pairs.append([tags[i][0], tags[i+1][0], tags[i+2][0]])

for phrase in pairs:
    print(phrase)
```

```python
# 4. Summarize the text using TextRank Algorithm

from sumy.summarizers.text_rank import TextRankSummarizer
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer 
```

```python
parser = PlaintextParser.from_string(data, Tokenizer('english'))
```

```python
text_rank_summarizer = TextRankSummarizer()

summary = text_rank_summarizer(parser.document, sentences_count=3)
```

```python
for sent in summary :
    print(sent, '\n')
```

```python
# 5. Summarize the text using extractive summarization

from sumy.summarizers.lsa import LsaSummarizer
lsa_summarizer = LsaSummarizer()
lsa_summary = lsa_summarizer(parser.document, sentences_count=3)
```

```python
for sent in lsa_summary :
    print(sent, '\n')
```
