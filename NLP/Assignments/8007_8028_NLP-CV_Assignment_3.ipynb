### Question 1

#### Read a word from user and print all meanings of it.

```python
from nltk.corpus import wordnet as wn
word = input("Enter a word: ")

synsets = wn.synsets(word)

for i in range(0,len(synsets)):
    print(f"{i+1}.{synsets[i].definition()}")
```

#### Read a word from user and print all noun meanings of it.

```python
word = input("Enter a word: ")

noun_synsets = wn.synsets(word, pos='n')

for i in range(0,len(noun_synsets)):
    print(f"{i+1}.{noun_synsets[i].definition()}")
```

### Question 2

```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
from nltk.chunk import RegexpParser
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.corpus import stopwords
import string
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Maharashtra'

headers = {
    'User-Agent': 'Custom-Script/1.9 (tushar@gmail.com)',
    'Accept-Language': 'en-US,en;q=0.5'
}
```

```python
response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
data = soup.get_text()
print(data)
```

```python
tokens = word_tokenize(data)
```

```python
noise = set(stopwords.words('english')).union(set(string.punctuation))

tokens = [token for token in tokens if token.lower() not in noise]

print(tokens)
tagged = pos_tag(tokens)
```

```python
# 1. Apply chuncking and show the chunks of type <VBD><DT>

grammar = "PAST_NP: {<VBD><DT>}"
parser = RegexpParser(grammar)
result = parser.parse(tagged)

print(result)
```

```python
# 2. Find all the named entities present in article
entities = ne_chunk(tagged)

for item in entities:
    if hasattr(item, "label"):
        print(item.label(), ":", " ".join(word for word, tag in item))
```

```python
# 3. Access all the verbs and apply the porter stemmer and lancaster stemmer
ps = PorterStemmer()
ls = LancasterStemmer()

verbs = [word for word, tag in tagged if tag.startswith('VB')]

for verb in verbs:
    
    print(f"Word :{verb:15} | P_stem : {ps.stem(verb):15}| L_stem: {ls.stem(verb)}")
```

```python
# 4. Access all the verbs in past tense and lemmatize them. Show only unique verbs.
lem = WordNetLemmatizer()

past_verbs = {
    lem.lemmatize(word, pos='v')
    for word, tag in tagged
    if tag == 'VBD'
}

for verb in sorted(past_verbs):
    print(verb)
```

```python
# 5. Read only the adjectives from the text and find their synonyms.
adjectives = [word for word, tag in tagged if tag.startswith('J')]

for adj in adjectives:
    synonyms = set()

    for syn in wordnet.synsets(adj, pos=wordnet.ADJ):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())

    print(f"{adj:15} --> {list(synonyms)[:5]}")
```

```python
# 6. Read all the verbs and print their antonyms.
verbs = [word for word, tag in tagged if tag.startswith('V')]

for verb in set(verbs):
    antonyms = set()

    for syn in wordnet.synsets(verb, pos=wordnet.VERB):
        for lemma in syn.lemmas():
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())

    if antonyms:
        print(f"{verb:15}-> {list(antonyms)}")
```
