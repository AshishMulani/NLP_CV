```python
import requests
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
```

```python
sent = 'They told that their ages are 23, 25 and 29 respectively.'
```

#### 1. Find the average of ages.

```python
ages=[int(i) for i in sent.split() if i.isdigit()]
ages
sum(ages)/len(ages)
```

####  2. Count total number of words starts with 't' / 'T'

```python
words = word_tokenize(sent)
words
```

```python
words_starting_with_t = [word for word in words if word.lower().startswith('t')]
print("Total words starting with 't'/'T':", len(words_starting_with_t))
```

#### 3. Count total punctuation symbols.

```python
import string
punctuation_count = 0
for char in sent:
    if char in string.punctuation:
        punctuation_count += 1

print("Total punctuation symbols:", punctuation_count)
```

#### 4. Print the words in capital letter which ends with 'd'

```python
words_ending_with_d = [word.upper() for word in words if word.lower().endswith('d')]
print("Words ending with 'd' in capital letters:", words_ending_with_d)
```

#### 5. Merge all the words and print (no punctuations)

```python
new_str = ''
for char in sent :
  if char not in string.punctuation:
    new_str += char
print(new_str)
```

## Question-2

#### Fetch the Wikipedia page for the National Flag of India

```python

url = "https://en.wikipedia.org/wiki/Flag_of_India"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
full_text = soup.get_text()
print(full_text)
```

#### 1. Count total number of Words

```python
total_words = len(full_text)
print(f"Total number of words: {total_words}")
```

####  2. Count percentage of punctuation marks




```python
punctuation_chars = 0
for char in full_text:
    if char in string.punctuation:
        punctuation_chars += 1

total_chars = len(full_text)

punctuation_percentage = (punctuation_chars / total_chars) * 100

print(f"Percentage of punctuation marks: {punctuation_percentage:.2f}%")
```


#### 3. Print total number of lines

```python
text_lines = [line.strip() for line in full_text.splitlines() if line.strip()]

total_lines = len(text_lines)
print(f"Total number of lines: {total_lines}")
```

#### 4. How many words are in title case?

```python
title = soup.find('title')
print(title)

title_text = title.get_text()
words = title_text.split()

title_case_words = [word for word in words if word.istitle()]
print("Title Case Words found:", title_case_words)

print(f"Total words in title case: {len(title_case_words)}")
```

#### 5. Remove numbers from the text

```python
cleaned_text = ""
for char in full_text:
    if not char.isdigit(): 
        cleaned_text += char

print(cleaned_text)
```
