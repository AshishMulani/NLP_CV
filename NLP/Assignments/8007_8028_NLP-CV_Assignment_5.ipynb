### Question 1

```python
# 1. Download the data from the link below using web scrapping: https://en.wikipedia.org/wiki/Mount_Everest

import nltk
from nltk import word_tokenize, pos_tag
from nltk.stem import WordNetLemmatizer
from bs4 import BeautifulSoup
import requests
from collections import Counter
```

```python
url = 'https://en.wikipedia.org/wiki/Mount_Everest'

headers = {
    'User-Agent': 'Custom-Script/1.9 (tushar@gmail.com)',
    'Accept-Language': 'en-US,en;q=0.5'
}
```

```python
response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
data = soup.get_text()
print(data)
```

```python
# Find the count of Total “Mount” in the text.

mount_count = data.count("Mount")
print(f"Total count of 'Mount': {mount_count}")
```

```python
# Save the extracted plain text contents in file mount.txt in upper case.

with open("mount.txt", "w", encoding="utf-8") as file:
    file.write(data.upper())

with open("mount.txt", 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# Print the title of this page.

page_title = soup.title.string if soup.title else "No title found"
print(f"Page Title: {page_title}")
```

```python
# 2. Read the following file https://www.london.ac.uk/sites/default/files/study-guides/introduction-to-natural-language-processing.pdf
```

```python
from pypdf import PdfReader

data = '../datasets/introduction-to-natural-language-processing.pdf'
```

```python
# Print only page number 12 and 13 from it.

combined_text = ""

reader = PdfReader(data)

# text from page 12 and 13
page_12 = reader.pages[11].extract_text()
page_13 = reader.pages[12].extract_text()

if page_12:
    combined_text += page_12 + "\n\n"
if page_13:
    combined_text += page_13 + "\n\n"

combined_text
```

```python
# Save this contents in a file nlp.txt

output_file = 'nlp.txt'
with open(output_file, 'w', encoding = 'utf-8') as f:
    f.write(combined_text.strip())
```

```python
with open(output_file, 'r', encoding='utf-8') as f:
    print(f.read())
```

### Question 3

```python
# 3. Read the file from the link givcen below: https://icml.cc/Conferences/2013/wp-content/uploads/2013/06/Machine-Learning-and-Natural-Language-Processing.docx

import docx

doc_path = '../datasets/Machine-Learning-and-Natural-Language-Processing.docx'
doc = docx.Document(doc_path)
```

```python
# Read and display first 5 paragraphs from it.

paragraph_count = 0
full_text = ""

for para in doc.paragraphs:
    full_text += para.text + "\n"
    
    if paragraph_count < 5:
        print(para.text)
        print()
        paragraph_count += 1
```

```python
# Count total ‘The’ present in the text

words = full_text.split()
the_count = 0

for word in words:
    cleaned_word = word.strip(".,!?()\"':;").lower()
    if cleaned_word == 'the':
        the_count += 1

print(f"Total 'The' present in the text: {the_count}")
```
