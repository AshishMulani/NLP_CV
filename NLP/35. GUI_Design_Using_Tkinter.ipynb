#### Adding Widgets

```python
top = Tk()
top.geometry('800x500')
top.title('My first application')

def square():
    num=int(e.get())**2
    print("Square is", num)

l = Label(top, text='Enter number:', font=('Arial',20))
b = Button(top, text='Click Here', font=('Arial',20), command=square)
e = Entry(top, font=('Arial',20))


l.pack()
e.pack()
b.pack()

top.mainloop()
```

```python
top = Tk()
top.geometry('800x500')
top.title('My first application')

def square():
    num=int(e.get())**2
    result.config(text="Square is " + str(num))

l = Label(top, text='Enter number:', font=('Arial',20))
b = Button(top, text='Click Here', font=('Arial',20), command=square)
e = Entry(top, font=('Arial',20))
result = Label(top, text='Square is', font=('Arial',20))

l.pack(); e.pack(); b.pack(); result.pack()

top.mainloop()
```

```python
top = Tk()
top.geometry('800x500')
top.title('My first application')

def square():
    num=int(e.get())**2
    result.config(text="Square is " + str(num))

l = Label(top, text='Enter number:', font=('Arial',20))
b = Button(top, text='Click Here', font=('Arial',20), command=square)
e = Entry(top, font=('Arial',20))
result = Label(top, text='Square is', font=('Arial',20))

l.place(x=100,y=50); e.place(x=100,y=120); b.place(x=100,y=190); result.place(x=100,y=260)

top.mainloop()
```
