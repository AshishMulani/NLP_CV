```python
words1 = 'working', 'worker', 'worked', 'workable'
words2 =  'bigger', 'biggest', 'big'
words3 = 'consulting', 'consultation', 'consulted', 'consultative'
words4 = 'player', 'played', 'players', 'playing', 'plays'
```

```python
from nltk.stem import  PorterStemmer, LancasterStemmer
```

```python
from nltk.tokenize import word_tokenize
import string
from nltk.corpus import stopwords
```

```python
ps = PorterStemmer()
```

```python
for word in words1:
    print(word, '->',ps.stem(word))
```

```python
for tup in [words1, words2, words3, words4]:
    for word in tup:
        print(word, '->',ps.stem(word))
    print("\n")
```

```python
ls = LancasterStemmer()
```

```python
for tup in [words1, words2, words3, words4]:
    for word in tup:
        print(word, '->',ls.stem(word))
    print("\n")
```

```python
#Remove stopwords, punctuation and suffixes and print the list of words
sent = "Hello friends! How are you? I like Python Programming but my friend likes to program in python"
```

```python
swords = stopwords.words("english")
```

```python
sent_list = word_tokenize(sent)
porter_stemmer = []
lan_stemmer = []
for word in sent_list:
    if word not in string.punctuation and word.lower() not in swords:
        porter_stemmer.append(ps.stem(word))
        lan_stemmer.append(ls.stem(word))
```

```python
print(porter_stemmer) #porter stemmer
print(lan_stemmer) #lancaster stemmer
```
