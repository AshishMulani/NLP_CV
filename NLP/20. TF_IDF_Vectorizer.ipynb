```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
sent1 = "It was a beautiful rainy day that made my day awesome."
sent2 = "We made it awesome by adding more flavors in it."
```

```python
tfidf = TfidfVectorizer()
```

```python
new = tfidf.fit_transform([sent1, sent2])
```

```python
df = pd.DataFrame(new.toarray(), columns=tfidf.get_feature_names_out())
df
```

### Read the file - mreviews.txt, remove the stopwords and create tf*idf vectorize result

```python
import os
```

```python
file_path = os.path.join("datasets", "mreviews.txt")
with open(file_path, "r", encoding="utf-8") as file:
    lines = [line.strip() for line in file if line.strip()]
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(lines)
```

```python
df_result = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
df_result
```

### TF*IDF Vectorizer by Reading File

```python
f = open('datasets/mreviews.txt')
lines = f.readlines()
```
