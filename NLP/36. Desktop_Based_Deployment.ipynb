```python
from tkinter import *
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import joblib
```

```python
ps = PorterStemmer()
# def clean_text(sent):
#     tokens1 = word_tokenize(sent)
#     tokens2 = [token for token in tokens1 if token.isalpha()]
#     tokens3 = [token.lower() for token in tokens2]
#     if token.lower() not in stopwords.words('english'):
#         tokens4 = [ps.stem(token) for token in tokens3]
#     return tokens4
def clean_text(sent):
    tokens1 = word_tokenize(sent)
    tokens2 = [token for token in tokens1 if token.isalpha()]
    tokens3 = [token.lower() for token in tokens2]
    
    # Corrected line: filter stopwords while iterating
    tokens4 = [token for token in tokens3 if token not in stopwords.words('english')]
    
    # Apply stemming to the filtered list
    final_tokens = [ps.stem(token) for token in tokens4]
    
    return final_tokens
```

```python
model = joblib.load('model.bin')
tfidf = joblib.load('preprocess.bin')

top = Tk()
top.geometry('800x500')
top.title('SPAM Classifier')

def classify():
    message = str(e.get())
    message_vect = tfidf.transform([message])
    pred = model.predict(message_vect)
    if pred[0] == 'ham':
        message = "NOT Spam"
    else:
        message = "Spam"
    result.config(text='Message is: ' + message)

l = Label(top, text='Enter the message:', font=('Arial', 20))
b = Button(top, text='Check Message', font=('Arial', 20), command=classify)
e = Entry(top, width=80, font=('Arial', 20))
result = Label(top, text='Message is', font=('Arial', 20))

l.pack(); e.pack()
b.pack(); result.pack()

top.mainloop()
```
