```python
from flask import Flask
app = Flask(__name__)

@app.route('/')
def hello_world():
    return ('Hello World')

@app.route('/hello')
def hello_world():
    return ('Welcome to Python Programming')

if __name__ == '__main__':
    app.run()
    # app.run(debug=True)  we don't have to always stop and re-run the code instead we can apply changes
```
