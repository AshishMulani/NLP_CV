```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5,0.25)
x
```

```python
y = x * 3.76 + 10.29
y
```

```python
plt.plot(x,y)
```

```python
x = np.arange(0,5,0.25)
```

```python
y = (x **3) * 3.76 +10.29
```

```python
plt.plot(x,y)
```

```python
# Log Growth Function

plt.plot(x, np.log(x))
```

```python
# Sigmoid Growth Function

x = np.arange(-5,5,0.25)
y= 1 / (1 + np.exp(-x))
plt.plot(x,y)
```
