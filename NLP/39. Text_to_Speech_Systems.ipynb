#### PyTTS

```python
import pyttsx3

engine = pyttsx3.init()
voices = engine.getProperty('voices')

# Loop through available voices to find a female one
for index, voice in enumerate(voices):
    print(f"Index: {index} | Name: {voice.name} | Gender: {voice.gender}")

# Usually, index 1 is the default Windows female voice (Microsoft Zira)
engine.setProperty('voice', voices[1].id) 

engine.say('''wait 
           a
           minute
           who
           are 
           you''')
engine.runAndWait()
```
